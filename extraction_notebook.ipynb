```python
from collections import Counter
from pydriller import Repository

authors = Counter()
for c in Repository('https://github.com/TheAlgorithms/Python').traverse_commits():
    #if 'bot' not in c.author.name:
    name = c.author.name if c.author and c.author.name else 'Unknown'
    authors[name] += 1

for author, count in authors.most_common():
    print(f"{author}: {count}")
```
